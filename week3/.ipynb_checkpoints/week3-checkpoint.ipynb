```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(0.01,16)
y = np.array([4800,4200,3400,3200,3001,2701,2660,2501,2420,2200,2000,1760,1723,1650,1400,1000])
Y = np.log(y)
X = np.stack((np.ones(len(x)),x)).T
data = pd.DataFrame([x,y,np.log(y)]).T

w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
f = w[0] + w[1]*x
b = np.exp(w[0])
a = w[1]
h = b*np.exp(a*x)
plt.scatter(x,y)
plt.plot(x,h,color='r')
print(w)
```
